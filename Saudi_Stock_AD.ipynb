----------------------------------------------------------------------------
No Trading Activity

كل الأربع شركات عندها نفس النمط:

NaN     NaN     NaN     ثابت     0       0

يعني:
لم يحدث أي تداول خلال اليوم، لذلك:
- لا يوجد سعر افتتاح (open)
- لا يوجد أعلى سعر (high)
- لا يوجد أقل سعر (low)

لكن:
- سعر الإغلاق (close) يبقى ثابت لأنه يمثل آخر سعر معروف قبل توقف التداول

📌 الخلاصة:

```python
print(f" اجمالي عدد الشركات: {df['trading_name '].nunique()}")
print("*"*100)

null_counts_T = df[df[['open','high','low']].isnull().all(axis=1)].groupby('trading_name ').size()

total_counts_T = df.groupby('trading_name ').size()

result = pd.DataFrame({
    'total': total_counts_T,
    'null_open_high_low': null_counts_T
}).fillna(0)

result['null_ratio'] = result['null_open_high_low'] / result['total']

#print(result.sort_values('null_open_high_low', ascending=False))

print(result.sort_values('null_open_high_low', ascending=False).to_string())
```

📌 الملاحظة:
القيم Null (35) تمثل عدد الأيام التي لا يوجد فيها نشاط تداول لهذه الشركات خلال الفترة.

❌ لذلك لا يمكن حذف هذه الصفوف، لأن الحذف سيعطي انطباعًا خاطئًا بأن الشركات غير موجودة في السوق خلال تلك الأيام.

❓ القرار:
هل يتم الاستبدال أم إنشاء Flag؟

⚠️ لماذا لا يتم الاستبدال (Mean Imputation)؟
- سيؤدي إلى توليد أسعار غير حقيقية
- يشوه البيانات المالية الفعلية
- غير منطقي في بيانات السوق لأن الأسعار لا تُخترع
- لو بنستبدل ٠ ، NaN ≠ 0 له معنى اخر

✔️ القرار النهائي:
تم اختيار إنشاء Flag بدلًا من الحذف أو الاستبدال.

📌 مبرر القرار:
بعد تحليل القيم المفقودة في الأعمدة السعرية (open, high, low)، تبيّن أن هذه القيم لم تكن ناتجة عن فقدان عشوائي، بل عن حالات "No Trading Activity"، حيث:
- volume_traded = 0
- no_trades = 0
- close = ثابت (آخر سعر معروف)

🔍 التصنيف:
Missing by Business Logic

أي أن القيم المفقودة ناتجة عن طبيعة السوق وليس خطأ في البيانات.

✔️ النتيجة:

تم الاحتفاظ بالبيانات كما هي، مع إضافة متغير (Flag) لتحديد أيام عدم وجود التداول،
لاستخدامه لاحقًا في التحليل دون تشويه البيانات الأصلية.

اضافه عامود No Trading Activity. = عشان برضو في الشارت استبعده  اذا احتجت




---




Missing Value Handling Decision




| Aspect                               | Mean Imputation                                                                                                                                                                                               | Flag Approach (Selected Method)                                                                                                                                        |
| ------------------------------------ | ------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------- | ---------------------------------------------------------------------------------------------------------------------------------------------------------------------- |
| Data Integrity                       | May create unrealistic price values                                                                                                                                                                           | Preserves original data                                                                                                                                                |
| Business Meaning                     | Ignores the trading context                                                                                                                                                                                   | Preserves the meaning of no trading activity                                                                                                                           |
| Financial Analysis Accuracy          | May distort financial results                                                                                                                                                                                 | Maintains analysis accuracy                                                                                                                                            |
| Impact on Models                     | May introduce bias into models                                                                                                                                                                                | Reduces misleading patterns                                                                                                                                            |
| Interpretability                     | Lower interpretability                                                                                                                                                                                        | Higher interpretability                                                                                                                                                |
| Performance                          | Requires additional computations to calculate replacement values and inject artificial data into the dataset, which may increase processing complexity and affect statistical calculations and model behavior | More efficient because it preserves the original data structure and only adds a lightweight Boolean flag (`True/False`) with minimal memory and computational overhead |
| Suitability for Financial Price Data | ❌ Not suitable                                                                                                                                                                                                | ✅ Suitable                                                                                                                                                             |
| Handling Type                        | Statistical Imputation                                                                                                                                                                                        | Context-Aware Handling                                                                                                                                                 |
| Final Decision                       | Not Selected                                                                                                                                                                                                  | ✅ Selected                                                                                                                                                             |

6-Activity Spike Analysis (Volume + Value)

حساب Z-Score للـ Volume و Value

هنا باخذها على مستويين ⁉

Company-level:

هل الشركة خرجت عن نمطها المعتاد؟

Sector-level:

هل الشركة تجاوزت سلوك قطاعها؟



 لو استخدمنا merge بنحتاج نسوي جدول جديد

 نحسب فيه mean و std لكل شركة،

وبعدها نرجع ندمجه مع الداتا الأصلية.

لكن باستخدام transform يطبق الحسابات مباشرة

 على كل شركة ويرجع النتائج بنفس عدد الصفوف،

 فصار أبسط وأنظف لحساب الـ Z-score.

```python
df["volume_z"] = df.groupby("company_name")["volume_traded"].transform( lambda x: (x - x.mean()) / x.std() )
```

```python
df["volume_spike"] = df["volume_z"] > 2
```

```python
df["volume_spike"].value_counts()
```

```python
volume_spike_companies = (
    df[df["volume_spike"] == True]
    .groupby("company_name")
    .size()
    .sort_values(ascending=False)
    .head(15)
)

volume_spike_companies
```

```python
import plotly.express as px

fig = px.box(
    df,
    y="volume_z",
    title="Volume Z-Score Distribution"
)

fig.update_layout(title_x=0.5)
fig.update_layout(template="plotly_dark")
```

```python

fig = px.bar(
    x=volume_spike_companies.index,
    y=volume_spike_companies.values,
    title="Top Companies with Volume Spikes",
    labels={
        "x": "Company",
        "y": "Spike Count"
    }
)

fig.update_layout(title_x=0.5)
fig.update_layout(template="plotly_dark")
```

```python
df["value_z"] = df.groupby("company_name")["value_traded"].transform(lambda x: (x - x.mean()) / x.std())

df["value_spike"] = df["value_z"] > 2
```
